```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data3 = pd.read_excel('comments.xlsx')
```

```python
data3
```

```python
data3.comment.unique()
```

```python
brod = data3[data3.comment.str.contains('@gabrielbrod')]
```

```python
brod
```

```python
felipexmachado = data3[data3.comment.str.contains('@felipexmachado')]
```

```python
gulartm = data3[data3.comment.str.contains('@gulartm')]
```

```python
label1 = []
for i in range(len(brod)):
    label1.append('@gabrielbrod')
brod['Atleta'] = label1

label2 = []
for i in range(len(felipexmachado)):
    label2.append('@felipexmachado')
felipexmachado['Atleta'] = label2

label3 = []
for i in range(len(gulartm)):
    label3.append('@gulartm')
gulartm['Atleta'] = label3
```

```python
df = pd.concat([brod, felipexmachado, gulartm], ignore_index = True)
cols = ["name", "Atleta"]
df = df[cols]
df
```

```python
df.Atleta.value_counts()
```

```python
plt.figure(2, figsize = (18, 10))

sns.set_style('darkgrid')

ax = sns.set_context("notebook", font_scale = 1.25, rc = {"lines.linewidth": 2.5})
ax = sns.barplot(y = df.Atleta.value_counts().index, 
                 x = df.Atleta.value_counts().values, 
                 palette = sns.color_palette("Purples_r", n_colors = 5), 
                 orient = 'h')
plt.xticks(rotation = 0, ha = 'right', fontsize = 18)
plt.yticks(fontsize = 18)
plt.xlabel('Contagem', fontsize = 20)
plt.ylabel('Atleta', fontsize = 20)
plt.title('Copa Podio Purple Belt', fontsize = 20)
plt.savefig('brod.png')
plt.show()
```
